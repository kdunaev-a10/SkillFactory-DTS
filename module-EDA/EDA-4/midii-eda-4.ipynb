```python
petersburg = [0.0974, 0.1352, 0.0817, 0.1016, 0.0968, 0.1064, 0.105]
magadan = [0.1033, 0.0915, 0.0781, 0.0685, 0.0677, 0.0697, 0.0764,
           0.0689]
```

# Проверка данных на нормальность

```python
H0 = 'Данные распределены нормально'
Ha = 'Данные не распределены нормально (мы отвергаем H0)'
alpha = 0.05
```

```python
_, p = shapiro(petersburg)
print('p=%.3f' % p)

# Интерпретация 

if p > alpha:
	print(H0)
else:
	print(Ha)
```

```python
_, p = shapiro(magadan)
print('p=%.3f' % p)

# Интерпретация 

if p > alpha:
	print(H0)
else:
	print(Ha)
```

```python
_, p = shapiro(magadan[:-1])
print('p=%.3f' % p)

# Интерпретация 

if p > alpha:
	print(H0)
else:
	print(Ha)
```

Данные в наборе 'petersburg' распределены нормально, а в наборе 'magadan' нет нормального распределения.

но если данные привести к одному размеру (убрать последнее значение из magadan) то оба становятся нормальными и можно провести проверку корреляции.

Так как первый признак это категория (местоположение), а второй признак это количественный из 2х разных групп то корреляцию можно провести с помощью "Независимый T-тест". 



```python
H0 = 'Нет значимой разницы между размером мидий в разных местах.'
Ha = 'Есть значимая разница между размером мидий в разных местах.'

midii_list = list(zip(petersburg, magadan))

midii_df = pd.DataFrame(midii_list, columns=['petersburg', 'magadan'] )
display(midii_df)
```

```python
def t_test(midii_df):
    print('\n' + "*** Результаты независимого T-теста ***")
    test_results = ttest_ind(midii_df['petersburg'], midii_df['magadan'], equal_var=True)

    p = round(test_results[1],2)

    if p>alpha:
        print(f"{p} > {alpha}. Мы не можем отвергнуть нулевую гипотезу. {H0}")
    else:
        print(f"{p} <= {alpha}. Мы отвергаем нулевую гипотезу. {Ha}")

t_test(midii_df)
```

Так как корреляция меньше установленно порога в 5%, то мы отвергаем нулевую гипотезу о одинаковом размере мидий в разных местах и принимаем гипотезу что они отличаются.

Дополнительно можно использовать днофакторный дисперсионный анализ ANOVA, который проверяет нулевую гипотезу о том, что две или более групп имеют одинаковое среднее значение генеральной совокупности. 

У нас 2 группы, так как требуются данные с нормальным распределением, то мы используем усеченый magadan 



```python
H0 = 'Нет значимой разницы между средним размером раковины мидий в двух разных местах.'
Ha = 'Есть значимая разница между средним размером раковины мидий в двух разных местах.'
_, p = f_oneway(petersburg, magadan[:-1])
print(_, p)
```

```python
if p>alpha:
  print(f"{p} > {alpha}. Мы не можем отвергнуть нулевую гипотезу. {H0}")
else:
  print(f"{p} <= {alpha}. Мы отвергаем нулевую гипотезу. {Ha}")
```

```python
print("разница в среднем размере мидии в зависимости от города-производителя ", round(midii_df['petersburg'].mean(0) - midii_df['magadan'].mean(0),2))
```

Вывод - гипотеза о разном размере мидий в разных места статистически значима.
