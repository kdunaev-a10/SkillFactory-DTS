В поле ниже введите, сколько дублирующихся записей вам удалось обнаружить. 

```python

wine_data_dup = wine_data.copy()



dupl_columns = list(wine_data_dup.columns)

mask = wine_data_dup.duplicated(subset=dupl_columns)
mask
wine_duplicates = wine_data_dup[mask]
print(f'Число оригинала: {wine_data_dup.shape[0]}')
print(f'Число дубликатов: {wine_duplicates.shape[0]}')
#
wine_dedupped = wine_data_dup.drop_duplicates(subset=dupl_columns)
print(f'Результирующее число записей: {wine_dedupped.shape[0]}')
```

```python
values = {
    'price': wine_dedupped['price'].mean().round(0)
}
#display(wine_dedupped_fill[wine_dedupped_fill['country'].isnull()])
null_colls_005 = [col for col in wine_dedupped.columns if (wine_dedupped[col].isnull().mean() > 0.05 and col != "price")]

for col in null_colls_005:
    values.update({col: "unknown"})
    
null_colls_001_005 = [col for col in wine_dedupped.columns if (wine_dedupped[col].isnull().mean() > 0 and wine_dedupped[col].isnull().mean() <= 0.05 and col != "price")]

for col in null_colls_001_005:
    values.update({col: wine_dedupped[col].mode()[0]})
print(values)

wine_dedupped_fill = wine_dedupped.fillna(value=values) 

display(wine_dedupped_fill.info())

sns.heatmap(wine_dedupped_fill.isnull()) # наглядная карта пропусков
```
